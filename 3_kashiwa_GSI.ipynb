# 看看柏市的政府相关数据

```python
import folium

# 1. 先创建一个没有默认底图的地图对象
m = folium.Map(
    location=[35.854, 139.968],  # 柏市周辺の緯度経度
    zoom_start=17,
    tiles=None  # これがポイント：デフォルトのOSMタイルを使わない
)

# 標準地図
folium.TileLayer(
    tiles="https://cyberjapandata.gsi.go.jp/xyz/std/{z}/{x}/{y}.png",
    attr="&copy; <a href='https://maps.gsi.go.jp/development/ichiran.html'>地理院タイル（国土地理院）</a>",
    name="GSI 標準地図",
    overlay=False,
    control=True
).add_to(m)

# 淡色地図
folium.TileLayer(
    tiles="https://cyberjapandata.gsi.go.jp/xyz/pale/{z}/{x}/{y}.png",
    attr="&copy; <a href='https://maps.gsi.go.jp/development/ichiran.html'>地理院タイル（国土地理院）</a>",
    name="GSI 淡色地図",
    overlay=False,
    control=True
).add_to(m)

# 空中写真
folium.TileLayer(
    tiles="https://cyberjapandata.gsi.go.jp/xyz/ort/{z}/{x}/{y}.png",
    attr="&copy; <a href='https://maps.gsi.go.jp/development/ichiran.html'>地理院タイル（国土地理院）</a>",
    name="GSI 空中写真",
    overlay=False,
    control=True
).add_to(m)


# 色別標高図
folium.TileLayer(
    tiles="https://cyberjapandata.gsi.go.jp/xyz/relief/{z}/{x}/{y}.png",
    attr="&copy; <a href='https://maps.gsi.go.jp/development/ichiran.html'>地理院タイル（国土地理院）</a>",
    name="GSI 色別標高図",
    overlay=False,
    control=True
).add_to(m)

# レイヤー切り替えコントロールを追加
folium.LayerControl().add_to(m)

m
```

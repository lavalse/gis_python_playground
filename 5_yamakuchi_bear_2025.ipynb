```python
import pandas as pd

df = pd.read_csv("data/r7-01011204.csv", encoding="SHIFT_JIS")  # 或 encoding="shift_jis"
df.head()
df.columns
```

```python
df = df.rename(columns={
    "追番": "id",
    "警察署": "police_station",
    "目撃(発見)年月日": "date",
    "目撃(発見)時間": "time",
    "目撃(発見)場所": "place",
    "状況": "situation",
    "頭数": "count",
    "体長": "length",
    "緯度": "lat",
    "経度": "lon",
})

df
```

```python
df.isna().mean().sort_values(ascending=False)
```

```python
df.loc[df["lat"] == 0, "lat"] = None
df.loc[df["lon"] == 0, "lon"] = None

df
```

```python
# 1. 删除整行空
df_clean = df.dropna(how='all')

# 2. 删除没有坐标的数据（用于地图）
df_geo = df_clean.dropna(subset=["lat", "lon"]).copy()

df_geo
```
